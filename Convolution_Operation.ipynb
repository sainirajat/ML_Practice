```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
im = cv2.imread('./img.jpg')
img = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.show()

print img.shape
```

```python
K = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])
print K.shape

print K
```

```python
out_r = img.shape[0] - K.shape[0] + 1
out_c = img.shape[1] - K.shape[1] + 1

new_img = np.zeros((out_r, out_c))
print img.shape, "--", K.shape
print new_img.shape
```

```python
for rx in range(out_r):
    for cx in range(out_c):
        image_patch = img[rx:rx+K.shape[0], cx:cx+K.shape[1]]
        prod = image_patch*K
        prod = prod.sum()
        new_img[rx, cx] = prod
```

```python
plt.imshow(new_img*(new_img>0), cmap='gray')
plt.show()
```
