```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
mean_01 = np.array([2.0,2.0])
cov_01 = np.array([[1.1,0.2],[0.2,1.1]])

data = np.random.multivariate_normal(mean_01,cov_01,650)
print data.shape
```

```python
plt.scatter(data[:,0],data[:,1],c='b')
plt.show()
```

```python
#d is the final shape Y
def pca(X,d=None):
    if d==None:
        d = X.shape[1]
    #covariance matrix
    A = np.cov(X,rowvar=False)
    
    #eigen values and eigen vector
    eig_vals,eig_vecs = np.linalg.eig(A)
    
    #projecting to new vector shape
    Z = X.dot(eig_vecs)
    
    #reduces dimension
    red_Z = Z[:,:d]
    
    return eig_vals,eig_vecs,Z,red_Z
```

```python
eig_vals,eig_vecs,Z,red_Z = pca(data,d=1)
print eig_vals.shape
print eig_vecs.shape
print red_Z.shape
print Z.shape
```

```python
#plotting reduced data
index = np.array(range(red_Z.shape[0]))
plt.scatter(index[:],red_Z[:,0],c='r')
plt.show()
```

```python
#now we will use eigen_vals
information = []
for ix in range(eig_vals.shape[0]):
    dx = [eig_vals[ix],eig_vecs[:,ix]]
    information.append(dx)
print information
```

```python
information = sorted(information,reverse=True,key=lambda z:z[0])
print information
```

```python
sorted_vals = []
sorted_vecs = []
for ix in range(eig_vals.shape[0]):
    sorted_vals.append(information[ix][0])
    sorted_vecs.append(information[ix][1])
sorted_vals = np.asarray(sorted_vals)
sorted_vecs = np.asarray(sorted_vecs)
print sorted_vals.shape
print sorted_vecs.shape
```

```python
explained_variance_ratio = sorted_vals/sum(sorted_vals)
print explained_variance_ratio
```
